```python
# 정답
# del 해주면 효율성 낮아져서 인덱스만 조절해줘야함
def solution(people, limit):
    answer = 0
    people=sorted(people, reverse=True)
    check = 0
    mini = len(people) -1
    while mini - check > 0:
        if limit >= people[check] + people[mini]:
            mini -= 1
        check += 1
        answer += 1
    if mini == check:
        answer+=1
    return answer
```
