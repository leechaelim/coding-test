```python
# 빡구현
def solution(numbers, hand):
    answer = ''
    left = [3,0]
    right = [3,2]
    for x in numbers:
        if x in (1, 4, 7):
            answer+="L"
            if x == 1:
                left = [0,0]
            if x == 4:
                left = [1,0]
            if x == 7:
                left = [2,0]
            
        elif x in (3, 6, 9):
            answer+="R"
            if x == 3:
                right = [0,2]
            if x == 6:
                right = [1,2]
            if x == 9:
                right = [2,2]
        else:
            
            if x == 2:#[0,1]
                l_distance = left[0] + abs(left[1]-1)
                r_distance = right[0] + abs(right[1]-1)
                if l_distance < r_distance:
                    answer+="L"
                    left = [0,1]
                elif l_distance > r_distance:
                    answer+="R"
                    right = [0,1]
                elif hand == "left":
                    answer+="L"
                    left = [0,1]
                else:
                    answer+="R"
                    right = [0,1]
            if x == 5:#[1,1]
                l_distance = abs(left[0]-1) + abs(left[1]-1)
                r_distance = abs(right[0]-1) + abs(right[1]-1)
                if l_distance < r_distance:
                    answer+="L"
                    left = [1,1]
                elif l_distance > r_distance:
                    answer+="R"
                    right = [1,1]
                elif hand == "left":
                    answer+="L"
                    left = [1,1]
                else:
                    answer+="R"
                    right = [1,1]
            if x == 8:#[2,1]
                l_distance = abs(left[0]-2) + abs(left[1]-1)
                r_distance = abs(right[0]-2) + abs(right[1]-1)
                if l_distance < r_distance:
                    answer+="L"
                    left = [2,1]
                elif l_distance > r_distance:
                    answer+="R"
                    right = [2,1]
                elif hand == "left":
                    answer+="L"
                    left = [2,1]
                else:
                    answer+="R"
                    right = [2,1]
            if x == 0:#[3,1]
                l_distance = abs(left[0]-3) + abs(left[1]-1)
                r_distance = abs(right[0]-3) + abs(right[1]-1)
                if l_distance < r_distance:
                    answer+="L"
                    left = [3,1]
                elif l_distance > r_distance:
                    answer+="R"
                    right = [3,1]
                elif hand == "left":
                    answer+="L"
                    left = [3,1]
                else:
                    answer+="R"
                    right = [3,1]
            
            
    return answer
```
