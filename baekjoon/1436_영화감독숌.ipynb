```python

# 내 코드 문제가 "0666" -> 666으로 저장해서 10666같은게 안생김 ㅜㅅㅜ
# 그래서 arr에 0666하나 더 넣어주고 arr copy에 int로 바꿔서 저장해주서 해결은 했는데, 그냥 완탐이 훨씬 좋다!
# 시간은 짧지만 메모리는 많은 코드
N = int(input())

arr = set(["666", "0666"])
arr_copy = set()
while len(arr_copy)<10001: #오 딱 최소로 구하긴 한다.
    tmp = set()
    for x in arr:
        for i in range(0,10):
            tmp.add(x+str(i))
            tmp.add(str(i)+x)
            arr_copy.add(int(x+str(i)))
            arr_copy.add(int(str(i)+x))
    arr |= tmp
arr_copy = list(arr_copy)
arr_copy.sort()
print(arr_copy[N-1])
```
