```python
# 내풀이 20분 쯤
from collections import deque
while True:
    string=input()
    if string=='.':
        break
    operation = deque() 
    check = True
    for idx,op in enumerate(string):
        if op=='(':
            operation.append("(")
        elif op==')':
            if len(operation)!= 0 and operation[-1] == "(":
                operation.pop()
            else:
                check = False
                break
        elif op=='[':
            operation.append("[")
        elif op==']':
            if len(operation)!= 0 and operation[-1] == "[":
                operation.pop()
            else:
                check = False
                break
    if check and len(operation)== 0:
        print('yes')
    else:
        print('no')
```

```python
#이지섭 풀이 반례: ([)()]).
def 괄호(a,b): #괄호들이 서로 짝을 이루었는지, () 가아니라 )( 이렇게 된 괄호는 없는지 확인
    
    if len(a)==len(b):
        for i,j in zip(a,b):
            if i>j:
                return False
        return True
    else:
        return False
def sm(string,a,b): #소괄호 안에 대괄호가 홀수개 있으면 불균형
    cnt=0
    for i,j in zip(a,sorted(b,reverse=True)):
        
        if '[' in string[i:j+1]:
            cnt+=1
        if ']' in string[i:j+1]:
            cnt+=1
            
    if cnt%2==0:
        return True

def big(string,a,b): #위와 반대
    cnt=0
    for i,j in zip(a,sorted(b,reverse=True)):
        
        if '(' in string[i:j+1]:
            cnt+=1
        if ')' in string[i:j+1]:
            cnt+=1
    if cnt%2==0:
        return True 

while 1:
    string=input()
    if string=='.':
        break
    print("스트링n",string)
    sm_left=[]
    sm_right=[]
    big_left=[]
    big_right=[]
    for idx,s in enumerate(string):
        if s=='(':
            sm_left.append(idx)
        elif s==')':
            sm_right.append(idx)
        elif s=='[':
            big_left.append(idx)
        elif s==']':
            big_right.append(idx)
    if 괄호(sm_left,sm_right) and 괄호(big_left,big_right):
        if sm(string,sm_left,sm_right) and big(string,big_left,big_right):
            print('yes')
        else:
            print('no')
    else:
        print('no')
```
