```python
#bfs
import collections

A, B = map(int, input().split())
if A>= B:
    print(A-B)
else:
    dp = [0]*(B+2)
    dp[A] = 0
    queue = collections.deque()
    queue.append(A)
    while queue:
        i = queue.popleft()
        if i==B:
            print(dp[B])
            break
        if i+1>=0 and i+1<B+2 and not dp[i+1]:#dp[i+1]이 0인경우만 값바꿔줌, 0 아니면 그게 최소니까
                dp[i+1] = dp[i] + 1
                queue.append(i+1)

        if i-1>=0 and i-1<B+2 and not dp[i-1]:
                dp[i-1] = dp[i] + 1
                queue.append(i-1)

        if i*2>0 and i*2<B+2 and not dp[i*2]:
                dp[2*i] = dp[i] + 1
                queue.append(2*i)
```

```python
# DP로 풀거다
def find_min(dp, i):
    INF = int(1e9)
    a,b,c,d = INF,INF,INF,INF
    if i-1>=0:
        if dp[i-1]!=0:
            a = dp[i-1]
    if i+1<len(dp):
        if dp[i+1]!=0:
            b = dp[i+1]
    if 2*i>=0 and 2*i<len(dp):
        if dp[2*i]!=0:
            c = dp[2*i]
    if i%2==0 and i<len(dp):
        if dp[i//2]!=0:
            d = dp[i//2]
    return min([a,b,c,d])
    
A, B = map(int, input().split())

dp = [0] * (2*B+1)
dp[A] = 1
for i in range(1,B):
    dp[i] = find_min(dp, i) + 1
print(dp[B]-1)
```

```python
from collections import deque 
def bfs(): 
    q = deque() 
    q.append(N) 
    while q: 
        v = q.popleft() 
        if v == K: 
            print(time[v]) 
            return 
        for next_step in (v-1, v+1, v*2): 
            if 0 <= next_step < MAX and not time[next_step]: 
                time[next_step] = time[v] + 1 
                q.append(next_step) 
MAX = 100001 
N, K = map(int, input().split()) 
time = [0]*MAX 
bfs()
```
